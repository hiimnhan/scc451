```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
import math
```

# Task 1: Basel Climate Dataset

```python
basel_df = pd.read_csv('ClimateDataBasel.csv')
```

## Data Preprocessing
### Data cleaning

```python
# OBSERVE DATA DISTRIBUTION
num_cols = basel_df.select_dtypes(include=['float64', 'int64']).columns
n_cols = 3  # number of plots per row
n_rows = math.ceil(len(num_cols) / n_cols)

plt.figure(figsize=(15, 4 * n_rows))

for i, col in enumerate(num_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.histplot(basel_df[col], kde=True, bins=30, color='lightcoral')
    plt.title(col)
    plt.tight_layout()

plt.show()
```
